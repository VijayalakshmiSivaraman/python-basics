# Bar Plot 

```python
f_data = pd.read_csv(r"C:\Users\lenovo\Documents\Uma\data\data fr viz\flight_delays.csv", index_col = 'Month')
```

```python
f_data.head()
```

```python
plt.figure(figsize = (16,6))
plt.title("Average arrival delay for Inceptez Airlines WN, by month")
sns.barplot(x = f_data.index, y = f_data['WN'])
plt.ylabel("Arrival delay in mins")
plt.xlabel("Month as integers")
```

```python
plt.figure(figsize = (16,6))
plt.title("Average arrival delay for Inceptez Airlines NK, by month")
sns.barplot(x = f_data.index, y = f_data['NK'])
plt.ylabel("Arrival delay in mins")
plt.xlabel("Month as integers")
```

# Heat map

```python
#tells magnitude of each location monthwise as colors
plt.figure(figsize = (8,8))
plt.title("Average arrival delay for Inceptez Airlines WN, by month")
sns.heatmap(data = f_data,annot = True)
plt.ylabel("Arrival delay in mins")
plt.xlabel("Month as integers")
```

```python
#tells magnitude 
plt.figure(figsize = (8,8))
plt.title("Average arrival delay for Inceptez Airlines WN, by month")
sns.heatmap(data = f_data)
plt.ylabel("Arrival delay in mins")
plt.xlabel("Month as integers")
```

```python
#tells magnitude 
plt.figure(figsize = (8,8))
plt.title("Average arrival delay for Inceptez Airlines WN, by month")
sns.heatmap(data = f_data,annot = True, vmin = -10, vmax = 40)
plt.ylabel("Arrival delay in mins")
plt.xlabel("Month as integers")
```

```python
#tells magnitude 
plt.figure(figsize = (8,8))
plt.title("Average arrival delay for Inceptez Airlines WN, by month")
sns.heatmap(data = f_data, cmap = 'YlGnBu')
plt.ylabel("Arrival delay in mins")
plt.xlabel("Month as integers")
```

```python
#tells magnitude 
plt.figure(figsize = (10,10))
plt.title("Average arrival delay for Inceptez Airlines WN, by month")
sns.heatmap(data = f_data,annot = True, vmin = -1, vmax = 1)
plt.ylabel("Arrival delay in mins")
plt.xlabel("Month as integers")
```

```python
f_data.corr()
```

```python
plt.figure(figsize = (10,10))
plt.title("Average arrival delay for Inceptez Airlines WN, by month")
sns.heatmap(data = f_data.corr(),annot = True, vmin = -1, vmax = 1)
plt.ylabel("Arrival delay in mins")
plt.xlabel("Month as integers")
```
