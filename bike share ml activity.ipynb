```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
m = model.fit(df[['temp']], df['registered'])
m
```

```python
Y = model.predict(df[['temp']])
Y
```

```python
plt.scatter(df[['temp']], df['registered'])
plt.plot(df[['temp']], Y, color = 'black')
plt.show()
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
r2 = r2_score(df['registered'], Y)
r2
```

```python
mae = mean_absolute_error(df['registered'], Y)
mae
```

```python
mse = mean_squared_error(df['registered'], Y)
mse
```

```python
rmse = np.sqrt(mse)
rmse
```

```python
mape = np.mean(np.abs((df['registered'] - Y)/df['registered'])) * 100
mape
```
