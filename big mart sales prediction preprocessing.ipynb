```python
numerical = df.select_dtypes(include = np.number).drop(columns= ['Item_Visibility', 'Outlet_Establishment_Year'])
numerical[0:2]
```

```python
categorical = df.select_dtypes(exclude = np.number)
categorical[0:2]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std_scale = StandardScaler()
numerical_preprocessing = std_scale.fit_transform(numerical)
numerical_preprocessing[0:2]
```

```python
categorical_preprocessing = pd.get_dummies(categorical)
categorical_preprocessing.head()
```

```python
numerical_preprocessed = pd.DataFrame(numerical_preprocessing, columns = numerical.columns)
numerical_preprocessed[0:2]
```

```python
preprocessed_df = pd.concat([numerical_preprocessed,categorical_preprocessing,df[['Item_Visibility', 'Outlet_Establishment_Year']]], axis = 1)
preprocessed_df.head()
```
